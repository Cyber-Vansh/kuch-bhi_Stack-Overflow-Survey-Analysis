# 03: EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed/cleaned_survey_results.csv')
```

```python
plt.hist(df['ConvertedCompYearly'], bins=50)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

```python
df_ed = df[df['EdLevel'] != 'Unknown']
df_ed.boxplot(column='ConvertedCompYearly', by='EdLevel', figsize=(12,6), rot=90)
plt.title('Salary by Education')
plt.suptitle('')
plt.show()
```

```python
plt.scatter(df['WorkExp'], df['ConvertedCompYearly'], alpha=0.3)
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
top_countries = df['Country'].value_counts().head(10)
top_countries.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
languages = df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack()
top_languages = languages.value_counts().head(10)

top_languages.plot(kind='bar', color='green', figsize=(10,6))
plt.title('Top 10 Most Popular Programming Languages')
plt.xlabel('Language')
plt.ylabel('Number of Developers')
plt.show()
```

```python
ai_counts = df['AISelect'].value_counts()
ai_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8,8))
plt.title('AI Tools Adoption Among Developers')
plt.ylabel('')
plt.show()
```

```python
df_sat = df[df['JobSat'] != 'Unknown']
df_sat.boxplot(column='ConvertedCompYearly', by='JobSat', figsize=(12,6), rot=45)
plt.title('Does Money Buy Happiness? (Salary by Job Satisfaction)')
plt.suptitle('')
plt.show()
```
