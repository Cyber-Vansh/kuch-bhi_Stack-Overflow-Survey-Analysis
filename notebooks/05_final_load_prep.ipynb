# 05: Final Load

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/cleaned_survey_results.csv')
```

```python
bins = [-1, 2, 5, 10, 20, 50]
labels = ['0-2 years', '3-5 years', '6-10 years', '11-20 years', '20+ years']
df['Experience_Level'] = pd.cut(df['WorkExp'], bins=bins, labels=labels)
```

```python
avg_salary = df.groupby('Country')['ConvertedCompYearly'].mean().reset_index()
avg_salary.rename(columns={'ConvertedCompYearly': 'Avg_Comp_by_Country'}, inplace=True)

df = df.merge(avg_salary, on='Country', how='left')
```

```python
df['Is_Data_Scientist'] = df['DevType'].str.contains('Data', na=False)

df['Primary_Language'] = df['LanguageHaveWorkedWith'].str.split(';').str[0]
df['Primary_Database'] = df['DatabaseHaveWorkedWith'].str.split(';').str[0]
```

```python
df.to_csv('../data/processed/tableau_load_data.csv', index=False)
```
