# 04: Statistics

```python
import pandas as pd
from scipy import stats
import statsmodels.api as sm
```

```python
df = pd.read_csv('../data/processed/cleaned_survey_results.csv')
```

```python
corr = df['WorkExp'].corr(df['ConvertedCompYearly'])
print(corr)
```

```python
remote_workers = df[df['RemoteWork'] == 'Remote']['ConvertedCompYearly']
in_person_workers = df[df['RemoteWork'] == 'In-person']['ConvertedCompYearly']

t_stat, p_val = stats.ttest_ind(remote_workers, in_person_workers, equal_var=False)

print(t_stat)
print(p_val)

if p_val < 0.05:
    print("Different salary")
else:
    print("Same salary")
```

```python
X = df['WorkExp']
y = df['ConvertedCompYearly']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
contingency_table = pd.crosstab(df['Age'], df['AISelect'])
print(contingency_table)

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square: {chi2}, P-value: {p}")

if p < 0.05:
    print("Age and AI usage are related")
else:
    print("Age and AI usage are not related")
```
